```python
from prometheus_api_client import PrometheusConnect
prom = PrometheusConnect(url ="http://20.71.55.156:9090", disable_ssl=True)

# Get the list of all the metrics that the Prometheus host scrapes
metrics = prom.all_metrics()
pprint.pprint(metrics)
```

```python
for metric in metrics:
    container_cpu_usage_seconds_total = prom.custom_query(query=metric)
    x = []
    for i in container_cpu_usage_seconds_total:
        x.append(float(i['value'][1]))

    if len(x) > 1:
        plt.plot(x)
        plt.ylabel(metric)
        plt.show()
```

```python
import datetime as dt
import pandas as pd
from prometheus_api_client import PrometheusConnect,  MetricSnapshotDataFrame, MetricRangeDataFrame



# metric values for a range of timestamps
metric_data = prom.get_metric_range_data(
    metric_name='go_memstats_alloc_bytes',
    start_time=(dt.datetime.now() - dt.timedelta(minutes=30)),
    end_time=dt.datetime.now(),
)

metric_df = MetricRangeDataFrame(metric_data)
metric_df.head()
print(len(metric_df))
print(len(metric_df[metric_df["instance"]=="cadvisor1"]))

print(set(list(metric_df["instance"])))
```

```python
from datetime import datetime

metric_df['value'] = pd.to_numeric(metric_df['value'])

cadvisor1 = metric_df[metric_df["instance"]=="cadvisor1"]

x = [datetime.fromtimestamp(time) for time in cadvisor1.index]
# pprint.pprint(x)
plt.plot(x, list(cadvisor1['value']))
# # plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')

for i in list(set(list(metric_df["instance"]))):
    y = metric_df[metric_df["instance"]==i]
    # pprint.pprint(x)
    y['value'] = pd.to_numeric(y['value'])

    y['value'].plot(label=i)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Go memory allocations")
plt.show()
#     plt.plot(x, list(y['value']))
    
```
